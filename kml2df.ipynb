```python
import pandas as pd
```

```python
import folium
```

```python
from bs4 import BeautifulSoup
```

```python
from random import randint
```

```python
sw_prefix = ['CB','KS','KZ','NS','YK']
```

```python
filename = 'kml/tundraSwan_USGS_ASC_argos_TUSW_CB1.kml'
```

```python
kml = open(filename).read()
```

```python
soup = BeautifulSoup(kml, 'xml')
```

```python
pms = soup.select('Placemark')
```

```python
sw_list = []

for b in sw_prefix:
    for i in range(1, 11):
        filename=f"kml/tundraSwan_USGS_ASC_argos_TUSW_{b}{i}.kml"
        R = randint(0, 255)
        G = randint(0, 255)
        B = randint(0, 255)
        color = f'#{R:02X}{G:02X}{B:02X}'
        kml = open(filename).read()
        soup = BeautifulSoup(kml, 'xml')
        pms = soup.select('Placemark')

        for pm in pms:
          try:
            sw = {}
            if 'TUSW' in pm.description.text:
              sw['id']=f"{b}{i}"
              sw['description'] = pm.description.text.strip()
              sw['longitude'] = float(pm.LookAt.longitude.text)
              sw['latitude'] = float(pm.LookAt.latitude.text)
              sw['range'] = float(pm.LookAt.range.text)
              sw['tilt'] = float(pm.LookAt.tilt.text)
              sw['heading'] = float(pm.LookAt.heading.text)
              sw['TimeStamp'] = pm.TimeStamp.text.strip()
              sw['coordinates'] = pm.Point.coordinates.text.strip()
              sw['color'] = color
              sw_list.append(sw)
          except Exception as e:
            print(e.args)
```

```python
df = pd.DataFrame(sw_list)
```
